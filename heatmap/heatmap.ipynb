```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

euro_df = pd.read_csv('../data/fully_cleaned_top_spotify_songs_EU.csv')
```

```python
country_avg = euro_df.groupby("country").mean(numeric_only=True)

country_avg
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(country_avg, annot=True, cmap="YlGnBu")
plt.title("Heatmap of Average Audio Features By Country")
plt.xlabel("Audio Features")
plt.ylabel("Country")
plt.show()
```
